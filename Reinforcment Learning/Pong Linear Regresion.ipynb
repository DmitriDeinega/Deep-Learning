```python
import torch
import gymnasium as gym
import random

import torch.nn as nn
import numpy as np

from torch.optim import SGD
from ale_py import ALEInterface
```

```python
env = gym.make("ALE/Pong-v5", full_action_space=False, obs_type='grayscale')
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
epsilon = 0.1
gamma = 0.999
num_episodes = 10000
```

```python
class PongModel(nn.Module):
    def __init__(self):
        super().__init__()

        self.linear = nn.Linear(33600, env.action_space.n)

    def forward(self, x):
        x = torch.flatten(x)
        x = self.linear(x)

        return x
```

```python
# model
model = PongModel()
model.to(device)

# loss
loss_fn = nn.MSELoss()

# Optimizer
optimizer = SGD(model.parameters(), lr=0.001, momentum=0.9)

for k in range(num_episodes):
    s, _ = env.reset()

    s_torch = torch.FloatTensor(s).to(device)
    
    done = False

    print('k:', k)

    while not done:
        q = model(s_torch)
        
        if random.random() >= epsilon:
            a = torch.argmax(q)
        else:
            a = env.action_space.sample()
            
        s_n, r, done, truncated, info = env.step(a) 
        
        s_n_torch = torch.FloatTensor(s_n).to(device)

        q_copy = q.clone()
        
        q_n = model(s_n_torch)
        
        q_copy[a] = r + gamma * torch.max(q_n)

        optimizer.zero_grad()
        
        loss = loss_fn(q, q_copy)
        loss.backward()
    
        optimizer.step()
        
        s_torch = s_n_torch
```
